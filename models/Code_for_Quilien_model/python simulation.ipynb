```python
class Simulator: 
    def __init__(self, s=0.1, formula='a & b & ~ c', ps=[0.1, 0.2, 1], rwv=[1, 0, 1]):
        self.s = s
        self.formula = parse_expr(formula)
        self.ps = ps 
        self.rwv = rwv

    def insert_vals(self, vals):
        '''
        This is badly written
        '''
        # Some disgusting manipulations to make sure that symbols stay in the right order
        input_ = dict(zip(list(sp.symbols(" ".join(map(str, np.sort(np.array(list(map(str,self.formula.atoms())))))))), vals))
        return int(bool(self.formula.subs(input_)))
    
    def simulate_s_(self, n):
        
        inps = []

        for _ in range(n):
            inp = []
            for k in range(len(self.rwv)):
                if np.random.choice(a=[False, True], p=[1-self.s, self.s]):
                    inp.append(bool(self.rwv[k]))
                else:
                    inp.append(np.random.choice(a=[False, True], p=[1-self.ps[k], self.ps[k]]))
            inps.append(inp)  
            
        out = []  
        inps_ = []  

        for inp_ in inps:
            out.append(self.insert_vals(inp_))
            inps_.append([int(v) for v in inp_])
            
        return inps_, out

    def simulate_(self, n):
        '''
        Rewrite as a boolean + random.binomial
        '''
        
        vars_ = []
        out = []

        for _ in range(n):
            vals = [np.random.choice(a=[False, True], p=[1-p_, p_]) for p_ in self.ps]
            vars_.append([int(v) for v in vals])
            out.append(self.insert_vals(vals))

        return vars_, out

    def event_importance(self, n, formulae='a & b'):

        MIs = []

        vars_, out = self.simulate_s_(n)

        for var in [_ for _ in zip(*vars_)]:
            MIs.append(np.corrcoef(var, out)[0, 1])

        return MIs, vars_, out
```
